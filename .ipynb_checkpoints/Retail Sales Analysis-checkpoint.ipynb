```python
## Retail Sales Data Analysis
"""
This notebook analyzes 12 months of sales data from an electronics store.
The analysis answers key business questions and provides actionable insights
through data visualization and statistical analysis.

Author: Zain Hammagi
Date: 09/11/2024
"""

# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import os
from itertools import combinations
from collections import Counter
import numpy as np
```

```python
def load_sales_data(data_path="./Sales_Data"):
    """
    Load and combine all sales data files from the specified directory.
    
    Parameters:
    -----------
    data_path : str
        Directory containing the sales data CSV files
        
    Returns:
    --------
    pandas.DataFrame
        Combined sales data from all CSV files
    """
    all_data = pd.DataFrame()
    files = [file for file in os.listdir(data_path) if not file.startswith('.')]
    
    for file in files:
        df = pd.read_csv(os.path.join(data_path, file))
        all_data = pd.concat([all_data, df], ignore_index=True)
    
    return all_data
```

```python
def clean_sales_data(df):
    """
    Clean the sales data by removing NaN values, fixing data types,
    and adding derived columns.
    
    Parameters:
    -----------
    df : pandas.DataFrame
        Raw sales data
        
    Returns:
    --------
    pandas.DataFrame
        Cleaned sales data with additional computed columns
    """
    # Remove rows with NaN values
    df = df.dropna(how='all')
    df = df[df['Order Date'].str[0:2] != 'Or']
    
    # Convert columns to correct data types
    df['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'])
    df['Price Each'] = pd.to_numeric(df['Price Each'])
    
    # Add Month column
    df['Month'] = pd.to_datetime(df['Order Date']).dt.month
    
    # Add Sales column
    df['Sales'] = df['Quantity Ordered'] * df['Price Each']
    
    # Add City column
    df['City'] = df['Purchase Address'].apply(lambda x: x.split(',')[1].strip() if isinstance(x, str) else '')
    
    # Add Hour column
    df['Hour'] = pd.to_datetime(df['Order Date']).dt.hour
    
    return df
```

```python
def analyze_monthly_sales(df):
    """
    Analyze monthly sales data and create visualization.
    
    Parameters:
    -----------
    df : pandas.DataFrame
        Cleaned sales data
    """
    results = df.groupby('Month')['Sales'].sum().reset_index()
    
    plt.figure(figsize=(10, 6))
    plt.bar(results['Month'], results['Sales'])
    plt.title('Monthly Sales Analysis')
    plt.xlabel('Month')
    plt.ylabel('Sales ($)')
    plt.grid(True, alpha=0.3)
    plt.xticks(range(1, 13))
    plt.show()
    
    return results

def analyze_city_sales(df):
    """
    Analyze sales by city and create visualization.
    
    Parameters:
    -----------
    df : pandas.DataFrame
        Cleaned sales data
    """
    results = df.groupby('City')['Sales'].sum().sort_values(ascending=False)
    
    plt.figure(figsize=(12, 6))
    results.plot(kind='bar')
    plt.title('Sales by City')
    plt.xlabel('City')
    plt.ylabel('Sales ($)')
    plt.xticks(rotation=45)
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
    
    return results
```

```python
def analyze_product_combinations(df):
    """
    Analyze which products are most commonly sold together.
    
    Parameters:
    -----------
    df : pandas.DataFrame
        Cleaned sales data
        
    Returns:
    --------
    Counter
        Most common product combinations
    """
    # Get orders with multiple products
    df_grouped = df.groupby('Order ID')['Product'].agg(list).reset_index()
    count_combinations = Counter()
    
    for products in df_grouped['Product']:
        if len(products) > 1:
            count_combinations.update(combinations(sorted(products), 2))
    
    return count_combinations.most_common(10)

def analyze_product_performance(df):
    """
    Analyze individual product performance.
    
    Parameters:
    -----------
    df : pandas.DataFrame
        Cleaned sales data
    """
    quantity_ordered = df.groupby('Product')['Quantity Ordered'].sum()
    prices = df.groupby('Product')['Price Each'].mean()
    
    fig, ax1 = plt.subplots(figsize=(12, 6))
    
    ax2 = ax1.twinx()
    ax1.bar(quantity_ordered.index, quantity_ordered, color='g', alpha=0.5)
    ax2.plot(prices.index, prices, color='b', linewidth=2)
    
    ax1.set_xlabel('Product')
    ax1.set_ylabel('Quantity Ordered', color='g')
    ax2.set_ylabel('Price ($)', color='b')
    
    plt.xticks(rotation=90)
    plt.title('Product Performance Analysis')
    plt.tight_layout()
    plt.show()
    
    return quantity_ordered, prices
```

```python
def main():
    # Load data
    print("Loading data...")
    sales_data = load_sales_data()
    
    # Clean data
    print("Cleaning data...")
    clean_data = clean_sales_data(sales_data)
    
    # Perform analysis
    print("\nAnalyzing monthly sales...")
    monthly_results = analyze_monthly_sales(clean_data)
    print("\nTop selling month:", monthly_results.loc[monthly_results['Sales'].idxmax(), 'Month'])
    
    print("\nAnalyzing city performance...")
    city_results = analyze_city_sales(clean_data)
    print("\nTop performing city:", city_results.index[0])
    
    print("\nAnalyzing product combinations...")
    combinations = analyze_product_combinations(clean_data)
    print("\nTop product combination:", combinations[0])
    
    print("\nAnalyzing product performance...")
    quantity, prices = analyze_product_performance(clean_data)
    
    return clean_data

if __name__ == "__main__":
    data = main()
```

```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import os

def load_sales_data(data_path="./Sales_Data"):
    """
    Load and combine all sales data files from the specified directory.
    """
    all_data = pd.DataFrame()
    files = [file for file in os.listdir(data_path) if not file.startswith('.')]
    
    for file in files:
        current_df = pd.read_csv(os.path.join(data_path, file))
        all_data = pd.concat([all_data, current_df], ignore_index=True)
    
    return all_data

def clean_sales_data(df):
    """
    Clean and prepare the sales data
    """
    # Drop rows with NaN values
    df = df.dropna(how='all')
    df = df[df['Order Date'].str[0:2] != 'Or']
    
    # Convert columns to correct data types
    df['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'])
    df['Price Each'] = pd.to_numeric(df['Price Each'])
    
    # Add Month column
    df['Month'] = pd.to_datetime(df['Order Date']).dt.month
    
    # Add Hour column
    df['Hour'] = pd.to_datetime(df['Order Date']).dt.hour
    
    # Add Sales column
    df['Sales'] = df['Quantity Ordered'] * df['Price Each']
    
    # Add City column
    df['City'] = df['Purchase Address'].apply(lambda x: x.split(',')[1].strip() if isinstance(x, str) else '')
    
    return df

# Load and prepare the data
df = load_sales_data()
df = clean_sales_data(df)

# Display the first 5 rows
print("First 5 rows of the cleaned DataFrame:")
print(df.head())

# Now you can proceed with the hourly analysis
def analyze_hourly_patterns(df):
    """
    Analyze sales patterns by hour of day
    """
    hourly_sales = df.groupby('Hour')['Sales'].sum()
    order_count = df.groupby('Hour')['Order ID'].count()
    
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8))
    
    # Sales by hour
    ax1.plot(hourly_sales.index, hourly_sales, marker='o')
    ax1.set_title('Sales by Hour of Day')
    ax1.set_xlabel('Hour')
    ax1.set_ylabel('Total Sales ($)')
    ax1.grid(True)
    
    # Order count by hour
    ax2.plot(order_count.index, order_count, marker='o', color='green')
    ax2.set_title('Number of Orders by Hour')
    ax2.set_xlabel('Hour')
    ax2.set_ylabel('Number of Orders')
    ax2.grid(True)
    
    plt.tight_layout()
    plt.show()
    
    # Find peak hours
    peak_sales_hour = hourly_sales.idxmax()
    peak_orders_hour = order_count.idxmax()
    
    print(f"""
    Peak Hours Analysis:
    - Highest sales occur at: {peak_sales_hour}:00
    - Most orders occur at: {peak_orders_hour}:00
    
    Recommended Advertising Times:
    - Primary: {peak_sales_hour-1}:00 - {peak_sales_hour+1}:00
    - Secondary: {peak_orders_hour-1}:00 - {peak_orders_hour+1}:00
    """)

# Run the hourly analysis
analyze_hourly_patterns(df)
```

```python
def detailed_peak_analysis(df):
    """
    Provides a comprehensive analysis of the peak hours with business insights
    """
    # Create more detailed hourly breakdown
    hourly_metrics = df.groupby('Hour').agg({
        'Sales': ['sum', 'mean', 'count'],
        'Quantity Ordered': ['sum', 'mean']
    }).round(2)
    
    # Evening Peak Analysis (19:00/7PM)
    evening_peak = hourly_metrics.loc[19]
    
    print("=== Detailed Peak Hour Analysis (19:00/7PM) ===")
    print(f"Total Sales: ${evening_peak['Sales']['sum']:,.2f}")
    print(f"Average Sale: ${evening_peak['Sales']['mean']:,.2f}")
    print(f"Number of Orders: {int(evening_peak['Sales']['count'])}")
    print(f"Total Items Sold: {int(evening_peak['Quantity Ordered']['sum'])}")
    print(f"Average Items per Order: {evening_peak['Quantity Ordered']['mean']:.1f}")
    
    # Analyze the surrounding hours (18:00-20:00)
    peak_window = hourly_metrics.loc[18:20]
    total_peak_sales = peak_window['Sales']['sum'].sum()
    total_peak_orders = peak_window['Sales']['count'].sum()
    
    print("\n=== Peak Window Analysis (18:00-20:00) ===")
    print(f"Total Window Sales: ${total_peak_sales:,.2f}")
    print(f"Total Window Orders: {int(total_peak_orders)}")
    print(f"Average Sales per Hour: ${total_peak_sales/3:,.2f}")
    
    # Calculate percentage of daily totals
    daily_sales = hourly_metrics['Sales']['sum'].sum()
    daily_orders = hourly_metrics['Sales']['count'].sum()
    
    peak_sales_percentage = (total_peak_sales / daily_sales) * 100
    peak_orders_percentage = (total_peak_orders / daily_orders) * 100
    
    print(f"\nPeak Window Represents:")
    print(f"- {peak_sales_percentage:.1f}% of daily sales")
    print(f"- {peak_orders_percentage:.1f}% of daily orders")
    
    print("\n=== Business Recommendations ===")
    print("1. Advertising Strategy:")
    print("   - Focus main advertising budget on 18:00-20:00 window")
    print("   - Start campaigns at 17:30 to capture early evening shoppers")
    print("   - Use targeted social media ads during this timeframe")
    
    print("\n2. Operations Optimization:")
    print("   - Ensure maximum staff availability during peak hours")
    print("   - Pre-stock popular items before peak window")
    print("   - Optimize website performance for higher traffic")
    
    print("\n3. Marketing Tactics:")
    print("   - Consider 'Evening Special' promotions")
    print("   - Implement flash sales during peak hours")
    print("   - Create dinner-time mobile notifications")
    
    # Visualize the peak window
    plt.figure(figsize=(12, 6))
    plt.plot(hourly_metrics.index, hourly_metrics['Sales']['sum'], 
             marker='o', color='blue', alpha=0.7)
    plt.fill_between([18, 19, 20], 
                    [0, 0, 0],
                    hourly_metrics['Sales']['sum'].loc[18:20],
                    alpha=0.3, color='green',
                    label='Peak Window')
    
    plt.title('Daily Sales Pattern with Peak Window Highlight')
    plt.xlabel('Hour of Day')
    plt.ylabel('Sales ($)')
    plt.grid(True, alpha=0.3)
    plt.legend()
    
    # Add peak window annotation
    plt.annotate('Peak Window\n18:00-20:00', 
                xy=(19, hourly_metrics['Sales']['sum'].loc[19]),
                xytext=(19, hourly_metrics['Sales']['sum'].loc[19]*1.1),
                ha='center',
                bbox=dict(boxstyle='round,pad=0.5', fc='yellow', alpha=0.5),
                arrowprops=dict(arrowstyle='->'))
    
    plt.tight_layout()
    plt.show()

# Run the detailed analysis
detailed_peak_analysis(df)
```
